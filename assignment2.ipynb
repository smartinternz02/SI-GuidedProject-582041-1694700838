3. Perform the Below Visualizations.
● Univariate Analysis
● Bi - Variate Analysis
● Multivariate Analysis

```python
# Univariate Analysis (Analysis on single feature 'living area' as that has the highest correlation with target Price)
sns.displot(df.living_area)
```

```python
# Bivariate Analysis (Comparision between 'living_area' feature and 'Price')

sns.jointplot(x='living_area',y='Price',data=df)
```

```python
# Multivariate analysis

imp_factors = ['living_area', 'grade_of_the_house', 'number_of_bathrooms', 'number_of_views', 'Area_of_the_basement', 'number_of_bedrooms']

sns.pairplot(df, vars=imp_factors + ['Price'])
plt.suptitle('Pair Plot of Numerical Variables vs. Price', y=1.02)
plt.show()
```
